```python
cd = covid.groupby(["departamento_nom"])[["edad"]].describe()
cd
```

```python
cd.columns
```

```python
cd.columns = cd.columns.droplevel(0)
```

```python
cd.columns
```

```python
# graficar

fig, ax = plt.subplots(figsize=(16,6))
ax.bar(x=cd.index, height=cd["count"], color="red", alpha=0.4)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# promedio de eddad por departamento
fig, ax = plt.subplots(figsize=(16,6))
ax.bar(x=cd.index, height=cd["mean"], color="green", alpha=0.4)
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
covid.columns
```

```python
for fecha in ['fecha_reporte_web', 'fecha_de_notificaci_n', 'fecha_inicio_sintomas', 'fecha_muerte', 'fecha_diagnostico', 'fecha_recuperado'] :
    covid[fecha] = pd.to_datetime(covid[fecha], format="%d/%m/%Y %H:%M:%S")
```

```python
covid.info()
```

```python
# agrupacion de por fechas

cf =  covid.groupby(["fecha_diagnostico"])[["edad"]].describe()
cf
```

```python
cf.columns = cf.columns.droplevel(0)
```

```python
cf.columns
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(cf.index, cf["count"], color="blue")
ax.set(title="Total de paciente diagnosticados cada dia",  xlabel="Fechas", ylabel="Numero de pacientes")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
cf["acumulados"] = cf["count"].cumsum()
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.plot(cf.index, cf["acumulados"], color="orange")
ax.set(title="Acumulados pacientes diagnosticados cada dia",  xlabel="Fechas", ylabel="Numero de pacientes")
ax.grid()
plt.xticks(rotation=45, ha="right")
plt.show()
```
